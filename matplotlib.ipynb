```python

plt.plot(x,y,label='sin',color='black')
plt.plot(x,z,label='cos',color='red')
plt.legend()
```

```python

plt.plot(x,y,label='sin',color='green',marker='o')
plt.plot(x,z,label='cos',color='orange')
plt.legend()
```

```python
fig=plt.figure(figsize=(15,5))
plt.plot(x,y,label='sin',color='green',marker='o')
plt.plot(x,z,label='cos',color='orange')
plt.legend()
plt.xlabel('THETA is in radians')
plt.ylabel('sin and cos')
```

```python
import matplotlib as mpl
```

```python
fig=plt.figure(figsize=(8,5),linewidth=3,edgecolor='b')
plt.plot(x,y)
ax=fig.axes[0]
ax.set_facecolor('#f1f1f1')
ax.set_xlabel('Theta')
ax.set_ylabel('sin(Theta)')
ax.xaxis.set_tick_params(direction='inout',color='r',length=5)
ax.xaxis.set_minor_locator(mpl.ticker.AutoMinorLocator(2))
ax.set_title('Axes')
ax.grid()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
fig.suptitle('Figure')
plt.show()
```

```python
import matplotlib as mpl
import seaborn as sns
```

```python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
fig.set_linewidth(1)
fig.set_edgecolor('b')
ax[0].plot(x,y)
ax[0].set_facecolor('#f1f1f1')
ax[0].set_xlabel('Theta')
ax[0].set_ylabel('sin(theta)')
ax[0].xaxis.set_tick_params(direction='inout',color='r',length=5)
ax[0].xaxis.set_minor_locator(mpl.ticker.AutoMinorLocator(2))
ax[0].set_title('Axes')
ax[0].grid()
ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)
fig.suptitle('Subplot 1 for axes=0')
plt.show()
```

```python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
fig.set_linewidth(4)
fig.set_edgecolor('b')
ax[1].plot(x,y)
ax[1].set_facecolor('#f1f1f1')
ax[1].set_xlabel('Theta')
ax[1].set_ylabel('cos(theta)')
ax[1].xaxis.set_tick_params(direction='inout',color='r',length=5)
ax[1].xaxis.set_minor_locator(mpl.ticker.AutoMinorLocator(2))
ax[1].set_title('Axes')
ax[1].grid()
ax[1].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)
fig.suptitle('Subplot 1 for axes=0')
plt.show()
```
